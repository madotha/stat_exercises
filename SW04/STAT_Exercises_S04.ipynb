## Aufgabe 4.1

**a)** Gegeben sind zwei unabhängige Zufallsvariable X und Y mit den Kennwerten

$\mu_X$ = 40, $\sigma_X$ = 15, 

$\mu_Y$ = 85 und $\sigma_Y$ = 18. 

Berechnen Sie $E(X +2Y)$, $Var(X + 2Y)$ und $E(X^2)$.

**Lösung:**

$$E(X + 2Y) = \mu_X + 2 \cdot \mu_Y = 40 + 2 \cdot 85 = 210$$

$$Var(X + 2Y) = \sigma^2_X + 2 \cdot \sigma^2_Y = 15^2 + 4 \cdot (18^2) = 1521$$

$$E(X^2) = Var(X) + (E(X))^2 = \sigma^2_X + \mu^2_X = 15^2 + 40^2 = 1825$$

**b)** Ein Werk produziert rechteckige Glasscheiben, deren Länge X und Breite Y (in
mmgemessen) voneinander unabhängig produktionsbedingten Schwankungen
unterliegen. Es gilt 

$\mu_X$ = 1000, $\sigma_X$ = 0.02, 

$\mu_Y$ = 500, $\sigma_Y$ = 0.01. 

Wie gross sind Erwartungswert und Standardabweichung des Umfangs $U$?

**Lösung:**

$$U = 2X + 2Y$$

$$E(U) = 2 \cdot E(X) + 2 \cdot E(Y) = 2 \mu_X + 2 \mu_Y = 3000$$

$$\sigma_U = \sqrt{4\sigma^2_X + 4\sigma^2_Y} = 0.0447$$

## Aufgabe 4.2

Erzeugen Sie mit *norm.rvs()* m = 500 Stichproben aus einer Standardnormalverteilung
mit Umfang n = 5. Speichern Sie die Stichproben als ($n \times m$)− Matrix ab.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
n = 5
m = 500
ran = np.array(norm.rvs(size=n*m))
sim = ran.reshape((n,m))
```

**a)** Stellen Sie die Stichproben als Runs graphisch dar, und staunen Sie über die zahlreichen möglichen Verläufe.

```python
plt.plot(sim)
plt.show()
```

**b)** Berechnen Sie nun für jede Stichprobe den Mittelwert $\bar{x}$, und erzeugen Sie ein
Histogramm der Mittelwerte. Aus der Theorie wissen Sie, dass die Mittelwerte normalverteilt mit Parameter $\mu$ = 0 und Standardabweichung $\sigma$ = $\frac{1}{\sqrt{n}}$ sein müssten. 

Überprüfen Sie dies graphisch mit Hilfe von Histogrammen und der
theoretischen Wahrscheinlichkeitsdichtekurve

```python
plt.hist(sim.T, bins=20, density=True, edgecolor="black", facecolor="white")
x = np.linspace(-4, 4, num=100)
y = norm.pdf(x)
plt.plot(x,y)
plt.title("Histogramm sim")
plt.show()

sim_mean = sim.mean(axis=0)
plt.hist(sim_mean, density=True, edgecolor="black",
facecolor="white")
x = np.linspace(-4, 4, num=100)
y = norm.pdf(x, loc=0, scale=1/np.sqrt(n))
plt.plot(x,y)
plt.title("Histogramm sim_mean")
plt.show()
```

## Aufgabe 4.3

**a)** Eine Elektronik-Firma stellt Widerstände her, die einen mittleren Widerstand
von 100$\Omega$ und eine Standardabweichung von 10$\Omega$ haben. Die Widerstände
sind normalverteilt. Bestimmen Sie die Wahrscheinlichkeit, dass sich für eine
zufällige Stichprobe von n = 25 Widerständen ein mittlerer Widerstand unter
95$\Omega$ ergibt.

$\overline{X}_n$ = mittlerer Widerstant in Stichprobe von $n$ Widerständen, ist normalverteilt

Mittelwert von $\mu_{\overline{X}_n} = \mu_X = 100\Omega$ 

Standardabweichung wäre dann $\sigma{\overline{X}_n} = \frac{\sigma_X}{\sqrt{n}} = \frac{10}{\sqrt{25}} = 2$

$\overline{X}_n \sim N(100, 4)$, wir suchen nun die Wahrscheinlichkeit $P(\overline{X}_n < 95)$:

```python
from scipy.stats import norm

norm.cdf(x=95, loc=100, scale=2)
```

**b)** Nehmen Sie an, dass eine Zufallsvariable $X$ einer uniformen Verteilung folgt,
und zwar $X \sim Uniform[4, 6]$. Wie lautet die Verteilung der Zufallsvariablen
$\bar{X_{40}}$, also des Mittelwertes einer Stichprobe vom Umfang $n$ = 40?

**Lösung:**

Mittelwert und Varianz von X lauten: 

$$\mu_X = 5$$

$$\sigma^2_X = \frac{(6-4)^2}{12} = \frac{1}{3}$$

Zentraler Grenzwertsatz deutet daraufhin, dass die Verteilung von $\overline{X}_{40}$ näherungsweise normalverteilt ist mit

$$Mittelwert: \mu_{\overline{X}_{40}} = 5$$

$$Varianz: \sigma_{\overline{X}_{40}} = \frac{\sigma_X}{\sqrt{n}} = \frac{1}{\sqrt{120}}$$

Folglich gilt also: $\overline{X}_{40} \sim N(5, \frac{1}{120})$

## Aufgabe 4.4

Es seien $\big\{ X_i \big\}_{1 \leq i \leq 50}$ unabhängige und normalverteilte Zufallsvariablen mit Erwartungswert
$\mu$ = 1 und Standardabweichung $\sigma$ = 2. Darüber hinaus sind folgende
Zufallsvariablen definiert:

$$ S_n = X_1 + X_2 + \cdots + X_n $$

$$ \overline{X}_n = \frac{1}{n} (X_1 + X_2 + \cdots + X_n) = \frac{S_n}{n}$$

Dabei ist $n$ = 50

**a)** Bestimmen Sie die Parameter der Normalverteilung von $S_n$ sowie $\overline{X}_n$.

**Lösung:**

Für $S_n$ gilt:

$$ E\big[ S_n \big] = E \bigg[ \sum_{i=1}^n X_i \bigg] = \sum_{i=1}^n X_i = 50 \cdot 1$$

$$ Var(S_n) = Var \bigg( \sum_{i=1}^n X_i \bigg) = \sum_{i=1}^n X_i Var(X_i) = 50 \cdot 4 = 200 $$

$S_{50}$ ist normal-verteilt mit $N(\mu_{S_{50}} = 50, \sigma_{S_{50}}^2 = 200)$. Für $\overline{X}_n$ gilt

$$ E[\overline{X}_n] \frac{1}{n} E \bigg[ \sum_{i=1}^n X_i \bigg] = 1$$

$$ Var(\overline{X}_n) = \frac{1}{n^2} Var \bigg( \sum_{i=1}^n X_i \bigg) = 0.08$$

$\overline{X}_50$ ist normal-verteilt mit $N(\mu_{\overline{X}_50} = 1, \sigma_{X_1}^2 = 0.08)$

**b)** Berechnen Sie die Wahrscheinlichkeit $P(E[X_1] − 1 \leq X_1 \leq E[X_1] + 1)$.

**Lösung:**

$X_1$ ist normal-verteilt mit $N(\mu_{X_{1}} = 1, \sigma_{X_{1}}^2 = 4)$. Wir suchen die Wahrscheinlichkeit $P(0 \leq X_1 \leq 2)$:

```python
from scipy.stats import norm

norm.cdf(x=2, loc=1, scale=2) - norm.cdf(x=0, loc=1, scale=2)
```

**c)** Berechnen Sie $P(E[S_n] − 1 ≤ S_n ≤ E[S_n] + 1)$.

**Lösung:**

Es gilt: $S_n \sim N(n \cdot \mu, n \cdot \sigma_{X}^2)$, also $S_{50} \sim N(50,200)$.

Wir suchen die Wahrscheinlichkeit $P(49 \leq S_{50} \leq 51)$:

```python
import numpy as np

norm.cdf(x=51, loc=50, scale=np.sqrt(200)) - norm.cdf(x=49, loc=50, scale=np.sqrt(200))
```

**d)** Berechnen Sie $P(E[\overline{X}_n] − 1 ≤ \overline{X}_n ≤ E[\overline{X}_n] + 1)$.

**Lösung:**

Es gilt: 

$$\overline{X}_n \sim N \Big(\mu, \frac{\sigma_{X}^2}{n} \Big)$$
$$\overline{X}_{50} \sim N \Big(1, \frac{4}{50}\Big)$$

Wir suchen die Wahrscheinlichkeit $P(0 \leq \overline{X}_{50} \leq 2)$:

```python
norm.cdf(x=2, loc=1, scale=np.sqrt(4/50)) - norm.cdf(x=0, loc=1, scale=np.sqrt(4/50))
```

## Aufgabe 4.5

Wir betrachten einen Datensatz, bei welchem zwei Methoden zur Bestimmung der
latenten Schmelzwärme von Eis verglichen werden. Wiederholte Messungen der
freigesetzten Wärme beim Übergang von Eis bei −0.7°C zu Wasser bei 0°C ergaben
die folgenden Werte (in cal/g):

![Tabelle der Methoden](4.5_table.png)

**a)** Berechnen Sie die arithmetischen Mittelwerte der beiden Methoden und geben
Sie für jede Methode den absoluten Fehler an.

**Absoluter Fehler für Methode A:**

```python
import numpy as np
from pandas import Series
methode_A = Series([79.98 , 80.04 , 80.02 , 80.04 , 80.03 , 80.03 , 80.04 , 79.97 , 80.05 , 80.03, 80.02 , 80.00 , 80.02])
# Artithmetisches Mittel
methode_A.mean()
```

Standardfehler $s_{\overline{X}_n} = \frac{s_X}{\sqrt{n}}$

```python
# Standardfehler
methode_A.std() / np.sqrt(methode_A.size)
```

Der Mittelwert mit absolutem Fehler für Methode A lautet demnach (80.02 $\pm$ 0.01)cal/g.

**Absoluter Fehler für Methode B:**

```python
methode_B = Series([80.02, 79.94, 79.98, 79.97, 80.03, 79.95, 79.97])
# Arithmetisches Mittel
methode_B.mean()
```

```python
#Standardfehler
methode_B.std() / np.sqrt(methode_B.size)
```

Der Mittelwert mit absolutem Fehler für Methode A lautet demnach (79.98 $\pm$ 0.01)cal/g.

**b)** Wie lauten die relativen Fehler?

Relativer Fehler: $\overline{x}_n \pm \frac{s_{\overline{x}_n}}{\overline{x}_n} \cdot 100\%$

**Relativer Fehler für Methode A:**

```python
methode_A.std() / np.sqrt(methode_A.size) / methode_A.mean()
```

Mit Umformen schreiben wir den Mittelwert mit relativem Fehler also 80.03 cal/g $\pm 8 \cdot 10^{-3}\%$

**Relativer Fehler für Methode B:**

```python
methode_B.std() / np.sqrt(methode_B.size) / methode_B.mean()
```
